**Create and visualize your model below...**

````r
rec <- recipe(mpg ~ displacement + cylinders, data = Auto) %>%
    step_ns(displacement, deg_free = 3) %>% 
    step_mutate(cyl_cat = factor(cylinders)) %>% 
    step_dummy(cyl_cat) %>% 
    step_rm(cylinders) 
````

````r
rec <- prep(rec)
````

````r
baked_data <- bake(rec, new_data = NULL)
````

````r
mod_fit <- linear_reg() %>%
  set_engine("lm") %>%
  fit(mpg ~ ., data = baked_data)
````

````r
predictions <- predict(mod_fit, new_data = baked_data) %>%
  bind_cols(Auto)

ggplot(predictions, aes(x = displacement, y = mpg)) +
  geom_point(size = 1, alpha = 0.35) +
  geom_line(aes(y = .pred, color = factor(cylinders)), linewidth = 1) +
  labs(title = "MPG vs Displacement",
       x = "Displacement",
       y = "MPG",
       color = "Cylinders")
````
